```python
# Importing necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading in data
raw_data  = pd. read_csv('KaggleV2-May-2016.csv', parse_dates=['ScheduledDay','AppointmentDay'])
raw_data.head(-5)
```

```python
# Creating a blank age dictionary up to the maximum age, then counting in the for loop and storing with appropriate age value
oldest_age=max(raw_data['Age'])
print(oldest_age)
count_age = dict.fromkeys(range(oldest_age+1),0)
#for loop to count and store values in dictionary
for a in range(oldest_age):
    for i in raw_data['Age']:
        if i == a:
            count_age[i] += 1
        
print (sum(count_age.values()))
```

```python
# Creating x and y values - our keys in our dict are the x and values are the y
x = list(count_age.keys())
y = list(count_age.values())

plt.plot(x,y)
plt.show()
```

```python
# Showing the number of appointments per day
x = raw_data['AppointmentDay'].value_counts()
plt.hist(x,bins=100)
plt.show()
```

```python
# Creating new column that indicates if a person had a pre-existing condition by if the value if greater than 0
raw_data['pre_exi_con']= raw_data['Hipertension'] + raw_data['Diabetes'] + raw_data['Alcoholism'] + raw_data['Handcap']
```

```python
# 
no_pec = raw_data['pre_exi_con'].value_counts()
print(no_pec)
count_no_pec = no_pec[0]
count_pec = sum (no_pec[1:])
```

```python
no_pec_no_show  = np.count_nonzero(np.logical_and(raw_data['pre_exi_con']== 0 ,raw_data['No-show']== 'Yes'))
pec_no_show = np.count_nonzero(np.logical_and(raw_data['pre_exi_con'] !=0 ,raw_data['No-show']== 'Yes'))
```

```python
plt.pie([count_no_pec,no_pec_no_show],labels = ['Show','No-show'],colors=['lightskyblue','red'],autopct='%1.2f%%')
plt.show()
plt.pie([count_pec,pec_no_show],labels = ['Show','No-show'],colors=['gold','red'],autopct='%1.2f%%')
plt.show()
```
